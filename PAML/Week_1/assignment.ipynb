```python
# How does the average watchtime differ between action and drama TV shows? Are there any other genres that stand out in terms of watchtime?
df['Watchtime'] = df['Watchtime'].astype(str).str.replace(',','').astype(float)
average_watchtime = df.groupby('Genre')['Watchtime'].mean()
average_watchtime[['Action', 'Drama']]
```

```python
# Heatmap: Create a heatmap to compare the average watch time for the top 10 genres. Display only the top 10 premiered years with the highest total watch time. 
grouped = df.groupby(['Genre', 'Premiere'])['Watchtime'].mean().reset_index()
top_genres = grouped.groupby('Genre')['Watchtime'].mean().nlargest(10).index
grouped = grouped[grouped['Genre'].isin(top_genres)]
top_years = grouped.groupby('Premiere')['Watchtime'].sum().nlargest(10).index
grouped = grouped[grouped['Premiere'].isin(top_years)]
pivot = grouped.pivot(index ='Genre', columns='Premiere', values='Watchtime')
plt.figure(figsize=(10, 8))
sns.heatmap(pivot, annot=True, fmt=".1f", cmap='viridis')
plt.title('Average Watch Time for Top 10 Genres and Premiered Years')
plt.show()
```

```python
# Bar plot: Visualize the distribution of watch times for the most-watched movies and TV shows. Plot a bar chart with bins representing different time intervals.
# How does the watch time vary across these intervals, and what insights can you draw from the distribution?
df['Watchtime'].plot(kind='hist',bins=10, color='skyblue')
plt.xlabel('Watchtime')
plt.ylabel('Frequency')
plt.title('Distribution of Watchtime')
plt.show()
# The distribution of watch time is right-skewed, with most of the watch times falling between 1000 and 2000 minutes. The number of movies and TV shows with watch times above 2000 minutes is relatively low.  
```
