```python
plt.plot(df['LoanAmount'],label = "Loan Amount",color="g")
plt.xlabel("Index")
plt.ylabel("Loan Amount Value")
plt.title("Plot of Loan Amount")

plt.legend()
plt.show()
```

```python
plt.plot(df['ApplicantIncome'],"g+",label = "Applicant Income")
plt.xlabel("Index")
plt.ylabel("Applicnt Income Value")
plt.title("Plot of Applocnst Income")

plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots()

legend_patches = [
    Patch(color='blue', label='Loan Amount'),
    Patch(color='green', label='Applicant Income'),
]

ax.plot(df["LoanAmount"],'b+', label="Loan Amount")
ax.plot(df["ApplicantIncome"],'g+', label = "Applicant Income")
ax.set_xlabel('time')
ax.set_ylabel('value')
ax.set_title("plot os loan ammount and applicant income")

plt.show()
```

```python
plt.scatter(df["ApplicantIncome"],df["LoanAmount"], color="blue", alpha=0.7, label = "Applicnt Income vs Loan Ammount")
plt.xlabel("Applicant Income")
plt.ylabel("Loan Amount")
plt.title("scatterplot of applicant income vs loan ammount")
plt.legend()
plt.show()
```
